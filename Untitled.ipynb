#### Create a DataFrame with all of the median ages from each neighborhood

```python
median_birthyear = pd.DataFrame(df.groupby('stop_nhbr')['birth year'].median())
median_birthyear.reset_index(inplace=True)
```

```python
median_birthyear.head(3)
```

Create one with out any 1969 years included

```python
median_birthyear_no_69 = pd.DataFrame(df[df['birth year'] != 1969 ].groupby('stop_nhbr')['birth year'].median())
median_birthyear_no_69.reset_index(inplace=True)
```

```python
median_birthyear_no_69.head(3)
```

Just from that we can tell that our the 69s were skewing our data.

```python
from citibike_scripts import change_birth_year
```

```python
df['update_birth_yr'] = df.apply(lambda row: change_birth_year(row,median_birthyear),axis=1)
df['update_birth_yr_no_69'] = df.apply(lambda row: change_birth_year(row,median_birthyear_no_69),axis=1)
```

```python
df
```

```python
start_nhbr = pd.read_csv('start_neighborhoods')
```

```python
df_w_snhbr = pd.merge(df, start_nhbr, on=['start station id'])
```

```python
df_w_snhbr = df_w_snhbr[['start_nhbr','usertype', 'gender',
       'start_month', 'start_day_of_week', 'start_hour', 'ave_temp', 'precip',
       'snow_depth', 'update_birth_yr','stop_nhbr']]
```

```python
df_means = df_w_snhbr.groupby('stop_nhbr').mean()
```

```python
df_means
```

```python
np.array(df_means.index).shape,
```

```python
fig = plt.figure(figsize=(20,10))
sns.heatmap(df_means.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap Between Mean Predictors', fontsize=20)
plt.tight_layout()
plt.savefig('corr_between_mean_preds_heatmap.png')
```

```python
fig = plt.figure(figsize=(20,10))
sns.heatmap(df_w_snhbr.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap Between Predictors', fontsize=20)
plt.tight_layout()
plt.savefig('corr_between_preds_heatmap.png')
```
